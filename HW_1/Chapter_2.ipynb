```python
import sympy as sp
import numpy as np
import math

__name__ = "ConvRate"

def main():
    #Variable
    a = sp.Symbol('a')

    #Step and lists
    step     = 10 ** (-3)
    q_list   = [0]
    a_list   = []

    for a_current in np.arange(-1.5, 5, step):
        char_root = [(-a - 1 - sp.sqrt(2 * a + 3)).evalf(subs={a:a_current}),
                     (-a - 1 + sp.sqrt(2 * a + 3)).evalf(subs={a:a_current})]
            
        tau = 2 / (char_root[0] + char_root[1])
        q   = 1 - tau * char_root[0]
        
        if (math.sqrt(max(abs(char_root[0]), abs(char_root[1]))) <= q) and (q < 1) and (q > q_list[-1]) and (q > 0): 
            q_list.append(q)
            q_list.sort()
            a_list.append(a_current)

    a_list.sort()
    print('Максимальное значение скорости q =', round(q_list[-1], 3), 'достигается при значении параметра a =', round(a_list[-1], 3))

if __name__ == "ConvRate":
    main()
```

```python
import numpy as np
import math
import random

__name__ = "PowerIter"

def Norm(vect):
    length = math.sqrt(vect[0]**2 + vect[1]**2 + vect[2]**2)
    vect = [value / length for value in vect]
    return vect

def Dot(vect1, vect2, dimension):
    result = 0
    for componient in range(0, dimension): result += vect1[componient] * vect2[componient]
    return result

def main():
    #Vectors
    x_prev = [random.randint(1,10) for i in range(0,3)]
    x_curr = [0, 0, 0]

    #Error value
    error  = 10**(-1)

    #Spectular radius
    spec_rad_real = 12.96
    spec_rad      = 0

    #Matrix
    matrix = [[0, -7, 7], [-7, -9, -5], [7, -5, -1]]

    while (abs(spec_rad - spec_rad_real) > error):
        for i in range(0, len(matrix)):
            for j in range(0, len(matrix)):
                x_curr[i] += round(matrix[i][j] * x_prev[j], 3)

        spec_rad = round(abs(Dot(x_curr, x_prev, len(matrix)) / Dot(x_prev, x_prev, len(matrix))), 3)
        x_curr = Norm(x_curr)
        
        for i in range(0, 3): 
            x_prev[i] = x_curr[i]
            x_curr[i] = 0
        
    print('Значение спекатрального радиуса матрицы A, ρ =', spec_rad)

if __name__ == "PowerIter":
    main()
```

```python
import numpy as np

__name__ = "GaussSeidel"

def ProblemMatrixFill(matrix, a, b, c, p, n):
    matrix[0][0] = b[0]
    for i in range(0, n + 1): matrix[-1][i] = p[i]
    for i in range(1, n):
        matrix[i][i]     = b[i]
        matrix[i][i - 1] = a[i]
        matrix[i][i + 1] = c[i]
    return matrix

def ProblemMatrix(a, b, c, p, n):
    matrix = [[0 for i in range(0, n + 1)] for j in range(0, n + 1)]
    matrix = ProblemMatrixFill(matrix, a, b, c, p, n)
    return matrix

def ProblemExtMatrix(a, b, c, f, p, n):
    matrix = [[0 for i in range(0, n + 2)] for j in range(0, n + 1)]
    matrix = ProblemMatrixFill(matrix, a, b, c, p, n)
    for k in range(0, n + 1): matrix[k][-1] = f[k]
    return matrix

def Selection_Major_Row(matrix, col, n):
    max_element = matrix[col][col]
    max_row = col
    for i in range(col + 1, n):
        if (abs(matrix[i][col]) > abs(max_element)):
            max_element = matrix[i][col]
            max_row = i
    
    if (max_row != col): matrix[col], matrix[max_row] = matrix[max_row], matrix[col]

def Solve_Gauss(matrix, n):
    for k in range(n - 1):
        Selection_Major_Row(matrix, k, n)
        for i in range(k + 1, n):
            matrix[i][-1] -= matrix[i][k] / matrix[k][k] * matrix[k][-1]
            for j in range(k, n):
                matrix[i][j] -= matrix[i][k] / matrix[k][k] * matrix[k][j]

    x = [0 for i in range(n)]
    for k in range(n - 1, -1, -1): x[k] = round((matrix[k][-1] - sum([matrix[k][j] * x[j] for j in range(k + 1, n)])) / matrix[k][k], 3)
    return x

def Solve_Seidel(a, b, c, f, p, n, error):
    x_0 = [0 for i in range(0, n)]
    x   = []

    for i in range(0, n): x.append(f[i] / b[i])
    x.append(f[-1]/p[-1])

    while (abs(x[n - 1] - x_0[n - 1]) >= error):
        for i in range(1, n): x[i] = round((f[i] - (a[i] * x[i - 1] + c[i] * x[i + 1])) / b[i], 3)
        x[n] = round((f[n] - sum(p[i] * x[i] for i in range(0, n))) / p[n], 3)
        for i in range(0, n): x_0[i] = x[i]
    return x

def FirstNorm(matrix, n):
    count_arr = [0 for j in range(0, n)]
    for i in range(0, n):
        for j in range(0, n):
            count_arr[i] += matrix[i][j]
        
    return max(count_arr)

def CondNumber(matrix, n): return FirstNorm(matrix, n) * FirstNorm(np.linalg.inv(matrix), n)

def main():
    #Input data
    n = 19

    a = [1 for i in range(0, n - 1)]
    a.insert(0, 0)

    b = [-2 for i in range(0, n - 1)]
    b.insert(0, 1)

    c = [1 for i in range(0, n - 1)]
    c.insert(0, 0)

    f = [round(2 / i**2, 3) for i in range(1, n)]
    f.insert(0, 1)
    f.insert(len(f), -6.333)

    p = [2 for i in range(0, n - 1)]
    p.insert(0, 1)
    p.insert(len(p), 1)

    #Solutions
    x1 = Solve_Seidel(a, b, c, f, p, n, 10**(-3))
    x2 = Solve_Gauss(ProblemExtMatrix(a, b, c, f, p, n), n + 1)

    #Characteristic roots
    char_root = np.linalg.eigh(ProblemMatrix(a, b, c, p, n))
    print('Минимальное собственное число: ', round(min(char_root[0]), 3), ', максимальное собственное число: ', round(max(char_root[0]), 3), sep = '')
    
    #Condition Number
    print('Число обусловленности:', CondNumber(ProblemMatrix(a, b, c, p, n), n))

    #Output solutions data
    for i in range(0, len(x1)): print('x_sei_', i, ' = ', x1[i], ', x_gau_', i, ' = ', x2[i], ', abs_error is ', round(abs(x2[i] - x1[i]), 3), sep='')

if __name__ == "GaussSeidel":
    main()
```

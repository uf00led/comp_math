```python
import numpy as np
import sympy as sp
import math

__name__ = "ApproxFuncU"

def Matrix_Minor(matrix, i, j): return [row[:j] + row[j + 1:] for row in (matrix[:i] + matrix[i + 1:])]

def Determinant(matrix):
    if len(matrix) == 2: return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
    determinant = 0
    determinant = sum((-1)**col * matrix[0][col] * Determinant(Matrix_Minor(matrix, 0, col)) for col in range(len(matrix)))
    return determinant

def Cramer(matrix, free_vector):
    det_matrix    = Determinant(matrix)
    det_solutions = []

    for i in range(0, len(matrix[0])):
        dup_matrix = [[matrix[i][j] for j in range(0, len(matrix[0]))] for i in range(0, len(matrix[0]))]
        for j in range(0, len(matrix[0])): dup_matrix[j][i] = free_vector[j]
        det_solutions.append(Determinant(dup_matrix))
        
    det_solutions = [value / det_matrix for value in det_solutions]
    return det_solutions

def main():
    x = sp.Symbol('x')
    y = sp.Symbol('y')

    x_prev = []
    func_1 = sp.tan(y - x) + x * y - 0.3
    func_2 = x * x + y * y - 1.5

    for x_current in np.arange(-math.sqrt(1.5), math.sqrt(1.5), 0.01):
        for y_current in np.arange(-math.sqrt(1.5), math.sqrt(1.5), 0.01):
            if (round(func_2.evalf(subs={x:x_current, y:y_current}), 1) == 0):
                if (round(func_1.evalf(subs={x:x_current, y:y_current}), 1) == 0):
                    x_prev.append([round(x_current, 2), round(y_current,2)])

    x_index = []
    for i in range(0,len(x_prev) - 1): 
        if (abs(x_prev[i][0] - x_prev[i + 1][0]) < 0.1): x_index.append(i)

    for i in range(0, len(x_index)): x_prev.pop(x_index[i] - i)

    func_1_diff = [sp.diff(func_1, x), sp.diff(func_1, y)]
    func_2_diff = [sp.diff(func_2, x), sp.diff(func_2, y)]
    free_vector = [-func_1, -func_2]

    x_curr = [[0, 0] for i in range(0, len(x_prev))]
    error = 10**(-6)

    for i in range(0, len(x_prev)):
        while True:
            matrix = [[func_1_diff[0].evalf(subs = {x:x_prev[i][0], y:x_prev[i][1]}),
                       func_1_diff[1].evalf(subs = {x:x_prev[i][0], y:x_prev[i][1]})],
                      [func_2_diff[0].evalf(subs = {x:x_prev[i][0], y:x_prev[i][1]}),
                       func_2_diff[1].evalf(subs = {x:x_prev[i][0], y:x_prev[i][1]})]]

            vector = [free_vector[0].evalf(subs = {x:x_prev[i][0], y:x_prev[i][1]}), 
                    free_vector[1].evalf(subs = {x:x_prev[i][0], y:x_prev[i][1]})]

            system_solutions = Cramer(matrix, vector)
            
            x_curr[i][0] = x_prev[i][0] + system_solutions[0]
            x_curr[i][1] = x_prev[i][1] + system_solutions[1]

            if (abs(x_prev[i][0] - x_curr[i][0]) < error): break
            x_prev[i][0] = x_curr[i][0]
            x_prev[i][1] = x_curr[i][1]

    print('Approximate solutions', x_curr)

if (__name__ == "ApproxFuncU"):
    main()
```

```python
import numpy as np
import sympy as sp
import math

__name__ = "ApproxNewton"

def RootCatcher(func, var):
    x_func = []
    for x_current in np.arange(-10, 10, 0.01):
        if (round(func.evalf(subs={var:x_current}), 1) == 0): x_func.append(round(x_current, 2))

    x_index = []
    for i in range(0,len(x_func) - 1): 
        if (abs(x_func[i] - x_func[i + 1]) < 0.1): x_index.append(i)

    for i in range(0, len(x_index)): x_func.pop(x_index[i] - i)

    print('Roots of equation: ', end = '')
    for i in range(0, len(x_func)): print("x", i, ' = ', x_func[i], ' ', sep = '', end = '')
    return x_func

def Newton(func, var, x_array, no_of_root, error):
    x_curr = 0
    x_prev = x_array[no_of_root - 1]

    while True:
        x_curr = x_prev - func.evalf(subs = {var:x_prev}) / sp.diff(func, var).evalf(subs = {var:x_prev})

        if (abs(x_curr - x_prev) < error): break
        else:                              x_prev = x_curr

    print("\nApproximate of", no_of_root ,"root of function", func, "is", x_curr)
    return x_curr

def main():
    x = sp.Symbol('x')
    func_1 = (0.5) ** x - (x - 1) ** 2 + 1
    func_2 = x**2 - sp.exp(x) / 5
    error = 10**(-10)

    Newton(func_1, x, RootCatcher(func_1, x), 1, error)
    Newton(func_2, x, RootCatcher(func_2, x), 1, error)

if (__name__ == "ApproxNewton"):
    main()
       
```
